```python
import pandas as pd
```

```python
data = pd.read_csv('adult.data.csv')
data.head()
```

**1. Сколько мужчин и женщин (признак *sex*) представлено в этом наборе данных?**

```python
male_num = data[data['sex'] == 'Male'].shape[0]
female_num = data[data['sex'] == 'Female'].shape[0]

print('{} мужчин'.format(male_num))
print('{} женщин'.format(female_num))
```

**2. Каков средний возраст (признак *age*) женщин?**

```python
female_mean_age = data[data['sex'] == 'Female']['age'].mean()

print('Средний возраст женщин {} лет'.format(female_mean_age))
```

**3. Какова доля граждан Германии (признак *native-country*)?**

```python
germans = data[data['native-country'] == 'Germany'].shape[0] / data.shape[0]

print('Доля граждан Германии {}'.format(germans))
```

**4-5. Каковы средние значения и среднеквадратичные отклонения возраста тех, кто получает более 50K в год (признак *salary*) и тех, кто получает менее 50K в год?**

```python
rich_mean_age = data[data['salary'] == '>50K']['age'].mean()
rich_disp = data[data['salary'] == '>50K']['age'].std()

not_rich_mean_age = data[data['salary'] == '<=50K']['age'].mean()
not_rich_disp = data[data['salary'] == '<=50K']['age'].std()

print('Среднее значение возраста людей с ЗП >50K: {}, среднеквадратичное отклонение: {}'.format(rich_mean_age, rich_disp))
print('Среднее значение возраста людей с ЗП <=50K: {}, среднеквадратичное отклонение: {}'.format(not_rich_mean_age, not_rich_disp))
```

**6. Правда ли, что люди, которые получают больше 50k, имеют как минимум высшее образование? (признак *education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters* или *Doctorate*)**

```python
data[data['salary'] == '>50K']['education'].value_counts()
```

Исходя из таблицы распределения признаков у группы людей с зарплатой выше 50K видно, что некоторые из них не имеют высшего образования

**7. Выведите статистику возраста для каждой расы (признак *race*) и каждого пола. Используйте *groupby* и *describe*. Найдите таким образом максимальный возраст мужчин расы *Amer-Indian-Eskimo*.**

**8. Среди кого больше доля зарабатывающих много (>50K): среди женатых или холостых мужчин (признак *marital-status*)? Женатыми считаем тех, у кого *marital-status* начинается с *Married* (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), остальных считаем холостыми.**

```python
married_list = ['Married-civ-spouse', 'Married-spouse-absent', 'Married-AF-spouse']

rich_married = data[(data['marital-status'].isin(married_list)) & (data['sex'] == 'Male') & (data['salary'] == '>50K')].shape[0] / data[(data['marital-status'].isin(married_list)) & (data['sex'] == 'Male')].shape[0]
rich_not_married = data[(~ data['marital-status'].isin(married_list)) & (data['sex'] == 'Male') & (data['salary'] == '>50K')].shape[0] / data[(~ data['marital-status'].isin(married_list)) & (data['sex'] == 'Male')].shape[0]
```

```python
print('Среди женатых мужчин доля зарабатывающих много {:.4f},\
а среди холостых {:.4f}'.format(rich_married, rich_not_married))
```

**9. Какое максимальное число часов человек работает в неделю (признак *hours-per-week*)? Сколько людей работают такое количество часов и каков среди них процент зарабатывающих много?**

```python
max_hours = data['hours-per-week'].max()
hard_workers_num = data[data['hours-per-week'] == max_hours].shape[0]
perc_hard_rich = data[(data['hours-per-week'] == max_hours) & (data['salary'] == '>50K')].shape[0] / hard_workers_num
```

```python
print('Максимальное число часов работы {} часов работают {} человек. \
Среди них доля зарабатывающих много {:.4f}'.format(max_hours, hard_workers_num, perc_hard_rich))
```
